# Step 2: Repeat the Hexagon Using a Nested Loop
Now that we have a hexagon using a loop, we can put this hexagon code inside another loop
that repeats, to get a number of hexagons arranged in a circle, each slightly overlapping each
other.
Modifying the pseudocode from Step 1, we can turn each hexagon just 10 degrees from the
previous one. To complete the circle, we need to do this a total of 360 ÷ 10 = 36 times (360 is
the total number of degrees in a circle).




```
Pseudocode
Repeat 36 times the following:
  Repeat 6 times the following:
    Move 100 steps forward
    Turn 60 degrees to the left
  Turn at 10 degrees to the right

```

**Notice how "Repeat" and the second "Turn" have the same level of indentation**

Remember, the comments after the # are optional. They only explain to the programmer what

````python
num_hexagons = 36
angle_turn = 360 / num_hexagons

for n in range(num_hexagons):
    for _ in range(6):
        shelly.forward(100)
        shelly.left(60)
    shelly.right(angle_turn)

shelly.hideturtle()
````

````python
import turtle

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
num_hexagons = 36
angle_turn = 360 / num_hexagons

shelly = turtle.Turtle()
turtle.bgcolor('black')

# Drawing the hexagon pattern
for n in range(num_hexagons):
    for i in range(6):
        shelly.color(colors[i])
        shelly.forward(100)
        shelly.left(60)
    shelly.right(angle_turn)

# Adding small white circles around the pattern
shelly.penup()
shelly.color('white')

for i in range(36):  # Repeat 36 times to match the 36 hexagons
    shelly.forward(220)
    shelly.pendown()
    shelly.circle(5)
    shelly.penup()
    shelly.backward(220)
    shelly.right(10)

shelly.hideturtle()
turtle.done()
````

# Experiment 1:
Start with this pseudocode to try creating a row of colored squares:


```
Pseudocode
Repeat 6 times the following:
  Set color from list
  Repeat 4 times the following:
    Move forward 20
    Turn left 90
  Put pen up
  Move forward 30
  Put pen down
Hide turtle

```




````python
import turtle

shelly = turtle.Turtle()

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
square_size = 20
gap_size = 30

# Drawing the row of colored squares
for color in colors:
    shelly.color(color)
    # Draw a square
    for _ in range(4):
        shelly.forward(square_size)
        shelly.left(90)
    # Move to the next position
    shelly.penup()
    shelly.forward(gap_size)
    shelly.pendown()

# Hide turtle when done drawing
shelly.hideturtle()

turtle.done()
````

# Experiment 3:
Can you make this green face using a series of circles? To help get you started here’s the code
for making one eye:


```

```




````python
import turtle
import math

turtle.bgcolor('honeydew')
turtle.setup(600, 600)
turtle.title("Refined Green Face")

shelly = turtle.Turtle()
shelly.speed(0)  # Fastest drawing speed

def draw_filled_circle(x, y, radius, color):
    shelly.penup()
    shelly.goto(x, y - radius)
    shelly.pendown()
    shelly.begin_fill()
    shelly.color(color)
    shelly.circle(radius)
    shelly.end_fill()

def draw_arc(x, y, radius, start_angle, end_angle, color, pensize=1):
    shelly.penup()
    shelly.goto(x, y)
    shelly.setheading(start_angle)
    shelly.pendown()
    shelly.color(color)
    shelly.pensize(pensize)
    shelly.circle(radius, end_angle - start_angle)

# Draw the face (big green circle with smoother gradient)
for i in range(100, 0, -1):
    green_value = 0.4 + (i / 250)  # Softer gradient
    color = (0.2, green_value, 0.2)  # Slightly muted green
    draw_filled_circle(0, 0, i, color)

# Draw the eyes
eye_radius = 30
pupil_radius = 12

# Left eye
draw_filled_circle(-45, 40, eye_radius, 'white')
draw_filled_circle(-45, 40, pupil_radius, 'forest green')
# Eyelid
draw_arc(-45, 40 + eye_radius, eye_radius, 180, 360, 'dark green', 2)

# Right eye
draw_filled_circle(45, 40, eye_radius, 'white')
draw_filled_circle(45, 40, pupil_radius, 'forest green')
# Eyelid
draw_arc(45, 40 + eye_radius, eye_radius, 180, 360, 'dark green', 2)

# Draw the nose (small triangle)
shelly.penup()
shelly.goto(-10, 5)
shelly.begin_fill()
shelly.color('lime green')
for _ in range(3):
    shelly.forward(20)
    shelly.left(120)
shelly.end_fill()

# Draw the mouth (gentle curve)
shelly.penup()
shelly.goto(-40, -20)
shelly.pendown()
shelly.color('light pink')
shelly.pensize(3)
shelly.setheading(-30)
shelly.circle(60, 60)

# Add rosy cheeks
cheek_color = '#FF9999'  # Light pink
draw_filled_circle(-70, -10, 20, cheek_color)
draw_filled_circle(70, -10, 20, cheek_color)

# Add eyebrows
shelly.penup()
shelly.color('dark green')
shelly.pensize(3)
shelly.goto(-70, 80)
shelly.setheading(-10)
shelly.pendown()
shelly.circle(80, 40)
shelly.penup()
shelly.goto(70, 80)
shelly.setheading(190)
shelly.pendown()
shelly.circle(-80, 40)

# Hide the turtle when done
shelly.hideturtle()

# Keep the window open until closed manually
turtle.done()
````

````python
import turtle
import random

turtle.bgcolor('blue')
shelly = turtle.Turtle()
shelly.speed(0)

colors = ["red", "orange", "yellow", "green", "purple", "pink", "white"]

shelly.penup()
shelly.goto(-50, -50)
shelly.pendown()
shelly.begin_fill()
shelly.color('gray')
for i in range(4):
    shelly.forward(100)
    shelly.left(90)
shelly.end_fill()

shelly.penup()
shelly.goto(-50, 50)
shelly.pendown()

shelly.begin_fill()
shelly.color('red')
shelly.left(60)
shelly.forward(100)
shelly.right(120)
shelly.forward(100)
shelly.right(120)
shelly.forward(100)
shelly.end_fill()

shelly.penup()
shelly.goto(-20, 10)
shelly.pendown()
shelly.begin_fill()
shelly.color('yellow')
for i in range(4):
    shelly.forward(40)
    shelly.left(90)
shelly.end_fill()

shelly.penup()
shelly.goto(0, -150)
shelly.pendown()

for _ in range(36):
    shelly.color(random.choice(colors))
    shelly.circle(100)
    shelly.right(10)

shelly.hideturtle()

turtle.done()
````

# Experiment 5:
Change the code at the end of Step 4 in your ​Create Geometric Art Project​ code to make
multiple circles coming back. Use the following:



```
Pseudocode
Repeat 36 times the following
    Lift pen
    Move forward 200
    Repeat 6 times
        Put pen down
        Make a circle of size 5
        Put pen up
        Move back 20
    Move back to center 80
    Turn 10 degrees to the right
Hide Turtle

```




````python
import turtle
import random

turtle.bgcolor('blue')
shelly = turtle.Turtle()
shelly.speed(0) 

colors = ["red", "orange", "yellow", "green", "purple", "pink", "white"]

shelly.penup()
shelly.goto(-50, -50)  
shelly.pendown()
shelly.begin_fill()  
shelly.color('gray')
for i in range(4):
    shelly.forward(100)
    shelly.left(90)
shelly.end_fill()  

shelly.penup()
shelly.goto(-50, 50)  
shelly.pendown()

shelly.begin_fill() 
shelly.color('red')
shelly.left(60)
shelly.forward(100)
shelly.right(120)
shelly.forward(100)
shelly.right(120)
shelly.forward(100)
shelly.end_fill() 

shelly.penup()
shelly.goto(-20, 10)  
shelly.pendown()
shelly.begin_fill() 
shelly.color('yellow')
for i in range(4):
    shelly.forward(40)
    shelly.left(90)
shelly.end_fill()  

shelly.penup()
shelly.goto(0, -150)  
shelly.setheading(0) 

for _ in range(36):
    shelly.penup()
    shelly.forward(200)
    for _ in range(6):
        shelly.pendown()
        shelly.color(random.choice(colors))  
        shelly.circle(5)
        shelly.penup()
        shelly.backward(20)
    shelly.goto(0, -150)  
    shelly.right(10)  

shelly.hideturtle()

turtle.done()
````
